# Connect Ollama as a local LLM

```python
# import libraries
import requests
from bs4 import BeautifulSoup
from IPython.display import Markdown, display
```

```python
# constant
OLLAMA_API = "http://localhost:11434/api/chat"
HEADERS = {"Content-Type":"application/json"}
MODEL = "llama3.2"
```

```python
messages = [
    {"role": "user",
    "content":"Describe some of business apllication of Generative AI, as a helpful agent."}
]
```

```python
payload = {
    "model":MODEL,
    "message":messages,
    "stream":False
}
```

```python
!ollama --version
```

```python
response = requests.post(OLLAMA_API, json=payload, headers=HEADERS)
print(response["message"]["content"])
```

```python
import ollama

response = ollama.chat(model=MODEL, messages=messages)
print(response['message']['content'])
```
